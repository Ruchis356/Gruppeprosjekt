```python
import requests
import pandas as pd

# Sett inn din klient-ID her
client_id = 'd933f861-70f3-4d0f-adc6-b1edb5978a9e'  # Bytt ut med din faktiske klient-ID

# Definer endpoint og parametere
endpoint = 'https://frost.met.no/observations/v0.jsonld'
parameters = {
    'sources': 'SN18700',  # Stasjons-ID for Trondheim (Værnes)
    'elements': 'air_temperature,precipitation_amount,wind_speed',  # Elementer du ønsker
    'referencetime': '2024-02-19/2024-02-25',  # Begrens tidsperioden for testing
    'timeresolutions': 'PT1H',  # Timevise data
}

# Send en HTTP GET-forespørsel
r = requests.get(endpoint, params=parameters, auth=(client_id, ''))

# Ekstraher JSON-data
json_data = r.json()

# Sjekk om forespørselen var vellykket
if r.status_code == 200:
    data = json_data['data']
    print('Data hentet fra frost.met.no!')
else:
    print('Feil! Statuskode:', r.status_code)
    print('Melding:', json_data['error']['message'])
    print('Årsak:', json_data['error']['reason'])
    exit()  # Avslutt hvis det er en feil

# Opprett en DataFrame for å lagre observasjonene
df = pd.DataFrame()

# Behandle dataene
for obs in data:
    row = pd.DataFrame(obs['observations'])
    row['referenceTime'] = obs['referenceTime']
    row['sourceId'] = obs['sourceId']
    df = pd.concat([df, row], ignore_index=True)

# Vis de første radene av DataFrame
print(df.head())
```
